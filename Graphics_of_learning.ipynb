```python
N = 2
acc_10_ep = (result['acc_10_ep'], result['val_acc_10_ep'])
acc_5_ep = (result['acc_5_ep'], result['val_acc_5_ep'])

ind = np.arange(N) 
width = 0.35       
plt.bar(ind, acc_10_ep, width, label='10 эпох')
plt.bar(ind + width, acc_5_ep, width,
    label='5 эпох')

# plt.ylabel('Scores')
plt.title('Доля правильных ответов при варьировании числа эпох')

plt.xticks(ind + width / 2, ('Обучающая', 'Тестовая'))
plt.xlabel("Тип выборки")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
x = ['10 эпох', '5 эпох']
energy = [result['time_to_learn_10_ep']/60, result['time_to_learn_5_ep']/60]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, energy, color='green')
plt.xlabel("Количество эпох")
plt.ylabel("Время, мин")
plt.title("Время обучения при варьировании числа эпох")

plt.xticks(x_pos, x)

plt.show()
```
