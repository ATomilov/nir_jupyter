```python
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(
            rect.get_x() + rect.get_width() / 2.0,
            1.02 * height,
            "%.3f" % height,
            ha="center",
            va="bottom",
            fontsize=12
        )
```
