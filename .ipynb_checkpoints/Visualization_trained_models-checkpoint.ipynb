```python
import numpy as np
import json
import os
import tensorflow as tf
from tensorflow.keras.utils import CustomObjectScope
from tensorflow.keras.initializers import glorot_uniform
from tensorflow.keras.models import Model, load_model, model_from_json
from tensorflow.keras.preprocessing import image
from sklearn.externals import joblib
import matplotlib.pyplot as plt
import cv2
import pickle
import seaborn as sns
import pandas as pd
```

```python
def get_models_paths(folder):
    files = os.listdir(folder)
    files.sort()
    files = ['{}/{}'.format(folder,file) for file in files if not file.startswith('.')]
    return files
```

```python
def prepare(filepath):
    IMG_SIZE = 100  # 50 in txt-based
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    img_array = img_array / 255
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
models_json = get_models_paths('trained_model_files/angles/JSON')
print('\n'.join('{}: {}'.format(*k) for k in enumerate(models_json)))
```

```python
models_h5 = get_models_paths('trained_model_files/angles/H5')
print('\n'.join('{}: {}'.format(*k) for k in enumerate(models_json)))
```

```python
json_file = open(models_json[32], "r")
loaded_model_json = json_file.read()
json_file.close()
loaded_model_CI = model_from_json(loaded_model_json)
loaded_model_CI.load_weights(models_h5[32])
loaded_model_CI.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
json_file = open(models_json[25], "r")
loaded_model_json = json_file.read()
json_file.close()
loaded_model_DI = model_from_json(loaded_model_json)
loaded_model_DI.load_weights(models_h5[25])
loaded_model_DI.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```
