```
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import os
from PIL import Image
import numpy as np

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

# First, load the image again
# dir_path = os.path.dirname(os.path.realpath(__file__))
filename = "4.bmp"
image = mpimg.imread(filename)
# image = Image.open(filename).convert('LA')
image_bin = Image.open(filename).convert('1')
image_bin.save('result.png')

grey_image = tf.image.rgb_to_grayscale(image)
# grey_image = tf.image.convert_image_dtype(image, dtype=tf.float32)
print(grey_image)
# plt.imshow(grey_image)
# plt.show()
# Create a TensorFlow Variable
x = tf.Variable(image, name='x')

model = tf.global_variables_initializer()

with tf.Session() as session:
  x = tf.transpose(x, perm=[0, 1, 2])
  session.run(model)
  result = session.run(x)

print(x.get_shape())
plt.imshow(result)
plt.show()

img = mpimg.imread(filename)     
gray = rgb2gray(img)    
plt.imshow(gray)
plt.show()

plt.imshow(image_bin)
plt.show()


# First, load the image again
# filename = "MarshOrchid.jpg"
# raw_image_data = mpimg.imread(filename)

# image = tf.placeholder("float", [None, None, 3])
# slice = tf.slice(image,[0,0,0],[-1,-1,1])

# with tf.Session() as session:
#     result = np.squeeze(result, 2)
#     plt.imshow(result)
#     plt.show()

# gray = tf.image.rgb_to_grayscale(image)
# print(gray)

# sess.run(tf.slice(gray, [0, 0, 0], [1, 3, 1]))
```
